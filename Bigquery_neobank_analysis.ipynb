```python
import pandas as pd
transaction_user_merge = pd.merge(df_users, df_transactions, on ='user_id', how="inner")
transaction_user_merge.head()
```

```python
# Does birth_year (age) affect transaction volume or total spend?
# Step1: To calculate the user's age and then categorize users into meaningful age brackets.
import pandas as pd
from datetime import datetime

# 1. Calculate Age
current_year = datetime.now().year
transaction_user_merge['age'] = current_year - transaction_user_merge['birth_year']

# 2. Define Age Bins
bins = [18, 25, 35, 45, 55, 65, 100] # Define the age ranges
labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']

# Create the new categorical column
transaction_user_merge['age_group'] = pd.cut(
    transaction_user_merge['age'],
    bins=bins,
    labels=labels,
    right=True, # Include the right edge of the bin (e.g., up to and including 25)
    include_lowest=True
)

print(transaction_user_merge[['birth_year', 'age', 'age_group']].head())
```

```python
# Step 2: Aggregate Transactions by Age Group
# Group the data by the new age_group column to calculate the required metrics:
# Transaction Volume: Use COUNT() on a transaction identifier (e.g., transaction_id).
# Total Spend: Use SUM() on the transaction amount column (e.g., amount).
age_analysis = transaction_user_merge.groupby('age_group').agg(
    # Transaction Volume (Count of unique transactions)
    num_transactions=('transaction_id', 'count'),

    # Total Spend (Sum of transaction amounts)
    total_spend=('amount_usd', 'sum')

).reset_index()

print(age_analysis)
```

```python
# Step 3: Visualize the Relationships. A. Transaction Volume vs. Age Group
import plotly.express as px

fig_volume = px.bar(
    age_analysis,
    x='age_group',
    y='num_transactions',
    color='age_group',
    title='Transaction Volume by Age Group',
    labels={
        "num_transactions": "Total Transaction Volume",
        "age_group": "Age Group"
    }
)
fig_volume.show()
```

```python
# B. Total Spend vs. Age Group
fig_spend = px.bar(
    age_analysis,
    x='age_group',
    y='total_spend',
    color='age_group',
    title='Total Spend Amount by Age Group',
    labels={
        "total_spend": "Total Spend (â‚¬)",
        "age_group": "Age Group"
    }
)
fig_spend.show()
```
